 ## IDENTIFY JOIN KEYS (CRITICAL)
- Purpose :- Confirm how household and person connect.

-  Shared columns (from your data):
   - FSU
   - Sample_Sg_Sb_No
   - Second_Stage_Stratum_No
   - Sample_Household_Number

- Conclusion
  - These four columns together uniquely identify a household.
  - This is our natural composite key.


```python
pd.read_sql("""
SELECT 
    FSU, Sample_Sg_Sb_No, Second_Stage_Stratum_No, Sample_Household_Number
FROM household
LIMIT 5;
""", conn)
```

```python
pd.read_sql("""
SELECT 
    FSU, Sample_Sg_Sb_No, Second_Stage_Stratum_No, Sample_Household_Number
FROM person
LIMIT 5;
""", conn)
```

Conclusion:
- The composite key (FSU, Sample_Sg_Sb_No, Second_Stage_Stratum_No, Sample_Household_Number) uniquely identifies households.
- The person table exhibits a one-to-many relationship with households, confirming that person-level data must be aggregated before joining to household data to avoid duplication and biased analysis.

- Required for foreign key modeling later

```python
pd.read_sql("""
SELECT
    FSU,
    Sample_Sg_Sb_No,
    Second_Stage_Stratum_No,
    Sample_Household_Number,
    COUNT(*) AS persons_per_household
FROM person
GROUP BY
    FSU,
    Sample_Sg_Sb_No,
    Second_Stage_Stratum_No,
    Sample_Household_Number
HAVING COUNT(*) > 1
LIMIT 10;
""", conn)
```

Conclusion:
- Person-level data shows multiple individuals per household, with household sizes ranging from 3 to 6 persons. This confirms a one-to-many relationship between households and persons, reinforcing the need to aggregate person-level attributes to the household level prior to any joins to avoid duplication and biased estimates.

## WEEKLY WORKING HOURS (JOB QUALITY)
- Purpose
  - Identify underemployment.

- Insight
  - Low totals → disguised unemployment
  - Wide variance → informal sector dominance

```python
pd.read_sql("""
SELECT
    AVG(
        IFNULL(Day1_Total_Hours,0) +
        IFNULL(Day2_Total_Hours,0) +
        IFNULL(Day3_Total_Hours,0) +
        IFNULL(Day4_Total_Hours,0) +
        IFNULL(Day5_Total_Hours,0) +
        IFNULL(Day6_Total_Hours,0) +
        IFNULL(Day7_Total_Hours,0)
    ) AS avg_weekly_hours_employed
FROM person
WHERE
    Age BETWEEN 15 AND 64
    AND CWS_Status_Code IN (11,12,21,31,41,51,61,71,81);
""",conn)
```

```python
pd.read_sql("""
SELECT
    CASE
        WHEN Age BETWEEN 15 AND 24 THEN 'Youth (15–24)'
        ELSE 'Adult (25–64)'
    END AS age_group,
    CASE
        WHEN (
            IFNULL(Day1_Total_Hours,0) +
            IFNULL(Day2_Total_Hours,0) +
            IFNULL(Day3_Total_Hours,0) +
            IFNULL(Day4_Total_Hours,0) +
            IFNULL(Day5_Total_Hours,0) +
            IFNULL(Day6_Total_Hours,0) +
            IFNULL(Day7_Total_Hours,0)
        ) < 35 THEN 'Underemployed'
        ELSE 'Adequately Employed'
    END AS job_quality,
    COUNT(*) AS person_count
FROM person
WHERE
    Age BETWEEN 15 AND 64
    AND CWS_Status_Code IN (11,12,21,31,41,51,61,71,81)
GROUP BY age_group, job_quality
ORDER BY age_group, person_count DESC;
""",conn)
```

## CREATE HOUSEHOLD DIMENSION TABLE
- Purpose :- One row per household
- Replace complex composite keys with household_id

- Why this design?
  - household_id → surrogate primary key
  - Natural keys retained → audit & debugging
  - Economic variables centralized → shared context



```python
cursor = conn.cursor()

cursor.execute("""
DROP TABLE IF EXISTS household_dim;
""")

conn.commit()
```

```python
cursor.execute("""CREATE TABLE household_dim AS
SELECT
    ROW_NUMBER() OVER (
        ORDER BY 
            FSU,
            Sample_Sg_Sb_No,
            Second_Stage_Stratum_No,
            Sample_Household_Number
    ) AS household_id,
               
    -- Survey info
    Panel,
    Quarter,
    Visit,
    Survey_Date,

    -- Natural household keys
    FSU,
    Sample_Sg_Sb_No,
    Second_Stage_Stratum_No,
    Sample_Household_Number,

    -- Geography
    Sector,
    State_Ut_Code,
    District_Code,
    NSS_Region,

    -- Household profile
    Household_Size,
    Household_Type,
    Religion,
    Social_Group,

    -- Economic indicators (standardized)
    Usual_Expenditure,
    Monthly_Consumer_Expenditure,
    Annual_Clothing_Expenditure / 12.0 AS Monthly_Clothing_Expenditure,
    Annual_Durables_Expenditure / 12.0 AS Monthly_Durables_Expenditure

FROM household;
""")
conn.commit()
```

## CREATE PERSON FACT TABLE (WITH FK)

- Purpose :- One row per person , Link every person to exactly one household


- Why this design?
  - person_id → unique person key
  - household_id → foreign key
  - Separates context (household) from outcomes (person)
  - This is textbook star-schema thinking.

```python
cursor.execute("""
CREATE TABLE person_fact AS
SELECT
     ROW_NUMBER() OVER (
        ORDER BY
            h.household_id,
            p.Person_Serial_No
    ) AS person_id,

    -- Foreign key
    h.household_id,

    -- Person identifiers
    p.Person_Serial_No,
    p.Relationship_To_Head,

    -- Demographics
    p.Sex,
    p.Age,
    p.Marital_Status,

    -- Education & skills
    p.General_Education_Level,
    p.Technical_Education_Level,
    p.Years_Formal_Education,
    p.Vocational_Training,

    -- Employment status
    p.Principal_Status_Code,
    p.Subsidiary_Status_Code,
    p.CWS_Status_Code,

    -- Industry & occupation
    p.Principal_Industry_Code,
    p.Principal_Occupation_Code,
    p.Subsidiary_Industry_Code,
    p.Subsidiary_Occupation_Code,

    -- Earnings
    p.CWS_Earnings_Salaried,
    p.CWS_Earnings_SelfEmployed,

    -- Weekly work (job quality)
    (
        IFNULL(p.Day1_Total_Hours,0) +
        IFNULL(p.Day2_Total_Hours,0) +
        IFNULL(p.Day3_Total_Hours,0) +
        IFNULL(p.Day4_Total_Hours,0) +
        IFNULL(p.Day5_Total_Hours,0) +
        IFNULL(p.Day6_Total_Hours,0) +
        IFNULL(p.Day7_Total_Hours,0)
    ) AS Weekly_Hours_Worked,

    -- Survey time dimensions
    p.Quarter,
    p.Visit

FROM person p
JOIN household_dim h
ON p.FSU = h.FSU
AND p.Sample_Sg_Sb_No = h.Sample_Sg_Sb_No
AND p.Second_Stage_Stratum_No = h.Second_Stage_Stratum_No
AND p.Sample_Household_Number = h.Sample_Household_Number;
""",)
conn.commit()
```

## CREATE THE FINAL ANALYTICAL FACT TABLE

```python
cursor = conn.cursor()
cursor.execute("""
CREATE TABLE employment_analytical_fact AS
SELECT
    -- Identifiers
    p.person_id,
    p.household_id,
               
    -- Demographics
    p.Age                         AS age,
    p.Sex                         AS sex,
    p.Marital_Status              AS marital_status,
               
    -- Education & skills         
    p.General_Education_Level     AS general_education_level,
    p.Technical_Education_Level   AS technical_education_level,
    p.Years_Formal_Education      AS years_formal_education,
    p.Vocational_Training         AS vocational_training,
    
    -- Employment status
            
    p.CWS_Status_Code             AS cws_status_code,
    p.Principal_Status_Code       AS principal_status_code,
    p.Subsidiary_Status_Code      AS subsidiary_status_code,
               
    -- Industry & occupation
    p.Principal_Industry_Code     AS principal_industry_code,
    p.Principal_Occupation_Code   AS principal_occupation_code,
               
    -- Earnings
    p.CWS_Earnings_Salaried       AS earnings_salaried,
    p.CWS_Earnings_SelfEmployed   AS earnings_self_employed,
    CASE
        WHEN p.CWS_Earnings_Salaried > 0 THEN 'Salaried'
        WHEN p.CWS_Earnings_SelfEmployed > 0 THEN 'Self-employed'
        ELSE 'No earnings'
    END AS earnings_type,
                          
    -- Job intensity
    p.weekly_hours_worked,
    p.underemployed_flag,
               
    -- Household context
    h.Sector                      AS sector,
    h.State_Ut_Code               AS state_ut_code,
    h.District_Code               AS district_code,
    h.Social_Group                AS social_group,
    h.Religion                    AS religion,
    h.Monthly_Consumer_Expenditure AS monthly_consumer_expenditure,
    h.Household_Size              AS household_size,

    -- Per-capita economic measure 
    (h.Monthly_Consumer_Expenditure * 1.0 / h.Household_Size)
        AS per_capita_mce,
           
    -- Time
    p.Quarter                     AS quarter,
    p.Visit                       AS visit
FROM person_fact p
JOIN household_dim h
ON p.household_id = h.household_id
               
--  Working-age restriction
WHERE p.Age BETWEEN 15 AND 64 ;
""")

conn.commit()
```

## Employment Analytical Fact Table:
- The employment_analytical_fact table is a person-level analytical dataset that integrates individual employment outcomes with household demographic and economic context. It includes key variables such as age, education, employment status, earnings type, weekly hours worked, and underemployment indicators, along with household characteristics like sector and per-capita consumption. The table is restricted to the working-age population (15–64) to ensure methodological correctness and is designed to support robust labour-market analysis across demographic, geographic, and socio-economic dimensions.
